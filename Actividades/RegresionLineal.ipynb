```python
# Cargar librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Cargar datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2

# Preprocesamiento de datos
X_train, X_test, y_train, y_test = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=100)

# Selección y entrenamiento del modelo
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Realizar predicciones
predicciones = modelo.predict(X_test)

# Calcular error cuadrático medio
mse = mean_squared_error(y_test, predicciones)
print("Error cuadrático medio:", mse)

# Graficar modelo de regresión lineal
plt.scatter(X_test, y_test, color='purple')
plt.plot(X_test, predicciones, color='green')
plt.title('Modelo de Regresión Lineal')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.show()
```
